# Data Analysis

## Gender

```python
# bar chart of Gender

ax = sns.countplot(x = 'Gender' ,data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data = df, x = 'Age', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Stress Level

```python
# plotting a bar chart for gender vs Stress Level

person_gen = df.groupby(['Gender'],as_index=False)['Stress Level'].sum().sort_values(by='Stress Level' , ascending=False)
sns.barplot(x = 'Gender',y= 'Stress Level' ,data = person_gen)
```

```python
ax = sns.countplot(data = df , x = 'Stress Level')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Stress Level vs Career

```python
# plotting a bar chart for Stress Level vs Career
sns.set(rc={'figure.figsize':(20,8)})
person_gen = df.groupby(['Occupation'],as_index=False)['Stress Level'].mean().sort_values(by='Stress Level' , ascending=False)
sns.barplot(x = 'Occupation',y= 'Stress Level' ,data = person_gen)
```

```python
ax = sns.countplot(data = df, x = 'Stress Level', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(data = df, x = 'Occupation', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Age vs Gender

```python
ax = sns.countplot(data = df, x = 'Age', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Physical Activity Level

```python
sns.set(rc={'figure.figsize':(20,8)})
person_pal = df.groupby(['Occupation'],as_index=False)['Physical Activity Level'].mean().sort_values(by='Physical Activity Level' , ascending=False)
sns.barplot(x = 'Occupation',y= 'Physical Activity Level' ,data = person_pal)
```

### Stress vs Gender

```python
ax = sns.countplot(data = df, x = 'Stress Level', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

### Heart Rate

```python

sns.set(rc={'figure.figsize':(20,8)})
person_pal = df.groupby(['Occupation'],as_index=False)['Heart Rate'].mean().sort_values(by='Heart Rate' , ascending=False)
sns.barplot(x = 'Occupation',y= 'Heart Rate' ,data = person_pal)
```

### Sleep Duration

```python
sns.set(rc={'figure.figsize':(20,8)})
person_sd = df.groupby(['Occupation'],as_index=False)['Sleep Duration'].mean().sort_values(by='Sleep Duration' , ascending=False)
sns.barplot(x = 'Occupation',y= 'Sleep Duration' ,data = person_sd)
```
